NMPC

- O NMPC busca a cada instante resolver um problema de otimização não linear respeitando as restrições impostas na dinâmica do sistema
- Modelo do Sistema
- Função Custo
- Loop de Controle

Biblioteca do_mpc

- Suporta NMPC completo
- Trabalha com sistemas não lineares
- Usa CasADi (otimização eficiente)

```python
import do_mpc
import numpy as np

model = do_mpc.model.Model('continuous')

x1 = model.set_variable('_x', 'x1')
x2 = model.set_variable('_x', 'x2')
u  = model.set_variable('_u', 'u')

model.set_rhs('x1', x2)
model.set_rhs('x2', -x1 + u)

model.setup()

mpc = do_mpc.controller.MPC(model)

setup_mpc = {
    'n_horizon': 20,
    't_step': 0.1,
}

mpc.set_param(**setup_mpc)

mterm = x1**2 + x2**2
lterm = x1**2 + x2**2 + u**2

mpc.set_objective(mterm=mterm, lterm=lterm)
mpc.setup()
```
